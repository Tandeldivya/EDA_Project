Now, in order to finalize our decision whether to drop these missing values or not we have to confirm if these missing values bring bias or not. That means we need to check the distribution of missing data. Whether it is MAR, NMAR or MCAR.

```python
msno.dendrogram(df)
plt.show()
```

As there is only one feature with missing values in them. The dendrogram will appear sparse or show a trivial pattern.

**So, in order to understand the nature of missingness. We observe our dataset and understand what each feature represents.**

In this dataset, **Electricity Installed Capacity (MW)** refers to the maximum amount of electricity that power generation facilities (like power plants) can produce at full operational capacity under ideal conditions. It is measured in megawatts (MW) and represents the total capacity of all electricity-generating equipment installed within a specific system, region, or facility.

So, this means that the missing values in this feature simply represent that the data was not available to be recorded for some countries for some years.

Hence, the data is missing due to intrinsic reasons because it might be related to specific features like country, technology, or year

Therefore, the distribution of these missing values is **Missing Not at Random (MNAR).**

So, our plan to handle these missing values is as follows:

1. Analyzing patterns of missingness:

  - Checking if missing values are concentrated to some specific groups like regions or years.

2. Handling missing values:

  - Adding a binary indicator column to retain information about missingness.

  - Using group-wise imputation based on the relevant features such as Region, Technology etc.


```python
# analyzing patterns of missingness
missing_distribution = df[df['Electricity Installed Capacity (MW)'].isnull()].groupby(['Region', 'Technology']).size()
```

```python
missing_distribution
```

```python
# adding a binary indicator column to retain information about missingness
df['Installed_Capacity_Missing'] = df['Electricity Installed Capacity (MW)'].isnull().astype(int)

# group-wise imputation based on the relevant features such as Region, Technology etc.
# here we use grouped median to impute
df['Electricity Installed Capacity (MW)'] = df.groupby(['Region', 'Technology'])['Electricity Installed Capacity (MW)'].transform(
    lambda x: x.fillna(x.median())
)

# verifying if missing values are handled or not
df.isnull().sum()
```

After imputation there are still some missing values left. Now, we analyze their pattern and try to handle them as well.

```python
remaining_missing = df[df['Electricity Installed Capacity (MW)'].isnull()]
remaining_missing
```

Imputing this remaining missing data with the overall median of the feature Electricity Installed Capacity (MW).

```python
df['Electricity Installed Capacity (MW)'].fillna(df['Electricity Installed Capacity (MW)'].median(), inplace=True)
```

```python
# verifying if there are any missing values still left
df.isnull().sum()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df['Year'] = pd.to_datetime(df['Year'], format='%Y')

df['Total_Generation'] = df['Electricity Generation (GWh)']
```

```python
def get_top_bottom_regions(df):
    total_generation = df.groupby('Region')['Total_Generation'].sum().sort_values(ascending=False)
    top_region = total_generation.index[0]
    bottom_region = total_generation.index[-1]
    
    top_generation = total_generation[0]
    bottom_generation = total_generation[-1]
    
    return top_region, top_generation, bottom_region, bottom_generation
```

```python
def get_top_bottom_technologies(df):
    tech_gen = df.groupby('Technology')['Total_Generation'].mean().sort_values(ascending=False)
    top_tech = tech_gen.index[0]
    bottom_tech = tech_gen.index[-1]
    
    top_avg_gen = tech_gen[0]
    bottom_avg_gen = tech_gen[-1]
    
    return top_tech, top_avg_gen, bottom_tech, bottom_avg_gen
```

4. Renewable Energy Technology Composition

```python
import matplotlib.pyplot as plt
import seaborn as sns


def plot_renewable_composition(df):
    renewable_df = df[df["RE or Non-RE"] == "Total Renewable"]

    renewable_tech_gen = renewable_df.groupby("Technology")["Total_Generation"].sum()

    renewable_tech_gen_sorted = renewable_tech_gen.sort_values(ascending=False)

    colors = sns.color_palette("husl", len(renewable_tech_gen_sorted))

    fig, (ax1, ax2) = plt.subplots(
        1, 2, figsize=(20, 10), gridspec_kw={"width_ratios": [2, 1]}
    )

    # Plot pie chart
    wedges, texts, autotexts = ax1.pie(
        renewable_tech_gen_sorted,
        labels=None,
        autopct=lambda pct: f"{pct:.2f}%" if pct > 0.5 else "",
        colors=colors,
        pctdistance=0.85,
        wedgeprops={"edgecolor": "white", "linewidth": 1},
    )

    plt.setp(autotexts, size=9, weight="bold", color="white")

    ax1.set_title(
        "Composition of Renewable Energy Technologies", fontsize=16, fontweight="bold"
    )

    legend_labels = [
        f"{tech} ({gen:,.0f} GWh)" for tech, gen in renewable_tech_gen_sorted.items()
    ]
    ax2.legend(
        wedges,
        legend_labels,
        title="Renewable Technologies",
        loc="center left",
        bbox_to_anchor=(0, 0.5),
    )
    ax2.axis("off")

    plt.tight_layout()
    plt.savefig(
        r"C:\Users\tandel divya\OneDrive\Desktop\Sem-5\EDA\Project\plots/renewable_technology_composition_with_legend.png",
        dpi=600,
        bbox_inches="tight",
    )
    plt.close()

    return renewable_tech_gen_sorted


# Call the function
renewable_composition = plot_renewable_composition(df)
print(renewable_composition)
```

```python
from sklearn.preprocessing import StandardScaler

# feature and target
X = df[['Electricity Installed Capacity (MW)']]
y = df[['Electricity Generation (GWh)']]

# initializing the scaler
scaler = StandardScaler()

# fit and transform the data (standardize)
X_scaled = scaler.fit_transform(X)
y_scaled = scaler.fit_transform(y)

# converting back to DataFrame for easier readability
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
y_scaled_df = pd.DataFrame(y_scaled, columns=y.columns)

print(X_scaled_df.head(), "\n\n", y_scaled_df.head())
```

```python
from sklearn.preprocessing import LabelEncoder
label_enc = LabelEncoder()
# create a copy of the data to work on
encoded_data = df.copy()

# loop through all columns and encode categorical ones
for col in encoded_data.select_dtypes(include=['object']).columns:
    encoded_data[col] = label_enc.fit_transform(encoded_data[col])
```

# Model Training

**Linear Regression**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# model creation
lin_model = LinearRegression()

# training the model
lin_model.fit(X_train, y_train)

# prediction
y_pred_lin = lin_model.predict(X_test)

# inverse transforming the predicted y and the testing y
y_pred_lin = y_pred_lin.reshape(-1,1)
y_test = np.array(y_test)
y_test = y_test.reshape(-1,1)
y_pred_lin = scaler.inverse_transform(y_pred_lin)
y_test = scaler.inverse_transform(y_test)

# calculate evaluation metrics
mse_lin = mean_squared_error(y_test, y_pred_lin)
rmse_lin = np.sqrt(mse_lin)  # Root Mean Squared Error
mae_lin = mean_absolute_error(y_test, y_pred_lin)
r2_lin = r2_score(y_test, y_pred_lin)  # R-squared

print(f"Linear Regression evaluation metrics - R²: {r2_lin}, MSE: {mse_lin}, RMSE: {rmse_lin}, MAE: {mae_lin}")
```
